```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, ax = plt.subplots()             # Create a figure containing a single Axes.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Plot some data on the Axes.
plt.show()                           # Show the figure.
```

```python
fig = plt.figure()             # an empty figure with no Axes
fig, ax = plt.subplots()       # a figure with a single Axes
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
# a figure with one Axes on the left, and two on the right:
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```

```python
f, ax = plt.subplots(2, 2)

f.set_size_inches(7 ,4)
f.set_facecolor('#eee')

ax[0, 0].plot(np.arange(0, 8))
ax[0, 0].grid()

ax[0, 1].plot(np.arange(8, 0, -1))
ax[0, 1].grid()
```

```python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(7,4))
gs = GridSpec(ncols=3, nrows=2, figure=fig)

ax1 = plt.subplot(gs[0,0])
ax1.plot(np.arange(6))
ax2 = fig.add_subplot(gs[1, 0:2])
ax2.plot(np.random.randint(low=0, high=10, size=(8)))
ax3 = fig.add_subplot(gs[:, 2])
ax3.plot(np.random.randint(low=0, high=10, size=(8)))
```

```python
from matplotlib.ticker import NullLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.arange(0, 6, 0.2))

ax.grid()
ax.xaxis.set_major_locator(NullLocator())
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.array([i**2 for i in range(5)]))

ax.grid()
ax.yaxis.set_major_locator(NullLocator())
```

```python
from matplotlib.ticker import LinearLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.array([np.sqrt(i) for i in range(17)]))

ax.grid()
ax.xaxis.set_major_locator(LinearLocator(5))
```

```python
from matplotlib.ticker import LinearLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.array([np.sqrt(i) for i in range(17)]))

ax.grid()
ax.yaxis.set_major_locator(LinearLocator(5))
```

```python
from matplotlib.ticker import MultipleLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.array([np.e ** i for i in range(17)]))

ax.grid()
ax.xaxis.set_major_locator(MultipleLocator(base=1))
```

```python
from matplotlib.ticker import FixedLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
ax.plot(np.array([np.e ** i for i in range(17)]))

ax.grid()
ax.xaxis.set_major_locator(FixedLocator([7, 8, 9, 10, 11 ,12, 13, 14, 15]))
```

```python
from matplotlib.ticker import NullLocator

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
ax.plot(x, np.sin(x))

ax.minorticks_on()
ax.grid(which='major', lw=2)
ax.grid(which='minor')

ax.xaxis.set_major_locator(NullLocator())
```

```python
from matplotlib.ticker import NullFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
ax.plot(x, np.sin(x))
ax.yaxis.set_major_formatter(NullFormatter())

ax.grid()
```

```python
from matplotlib.ticker import FormatStrFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
ax.plot(x, np.sin(x))
ax.yaxis.set_major_formatter(FormatStrFormatter('y = %.2f'))
ax.xaxis.set_major_formatter(FormatStrFormatter('x = %.2f'))

ax.grid()
```

```python
from matplotlib.ticker import FuncFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
ax.plot(x, np.sin(x))
ax.yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'[{x}]' if x > 0 else f'({x})'))
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, pos: f'[{x}]' if x > 0 else f'({x})'))

ax.grid()
```

```python
from matplotlib.ticker import ScalarFormatter

fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot()
x = np.arange(-2*np.pi, 2*np.pi, 0.1)
ax.plot(x, np.sin(x) * 10e4)
sf = ScalarFormatter()
sf.set_powerlimits((-4, 4)) #  10e4 <= y <= 10e-4
ax.yaxis.set_major_formatter(sf)


ax.grid()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(6, 10))
ax_3d = fig.add_subplot(projection='3d')

x = np.arange(-2*np.pi, 2*np.pi, 0.2)
y = np.arange(-2*np.pi, 2*np.pi, 0.2)
xgrid, ygrid = np.meshgrid(x, y)
zgrid = np.sin(xgrid) * np.sin(ygrid) / (xgrid * ygrid)

ax_3d.plot_wireframe(xgrid, ygrid, zgrid, color='r')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(6, 10))
ax_3d = fig.add_subplot(projection='3d')

x = np.arange(-2*np.pi, 2*np.pi, 0.2)
y = np.arange(-2*np.pi, 2*np.pi, 0.2)
xgrid, ygrid = np.meshgrid(x, y)
zgrid = np.sin(xgrid) * np.sin(ygrid) / (xgrid * ygrid)

ax_3d.plot_surface(xgrid, ygrid, zgrid, cmap='magma')
plt.show()
```
